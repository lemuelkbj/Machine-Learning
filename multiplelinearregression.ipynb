```python
import pandas as pd
import matplotlib.pyplot as mb
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
ds = pd.read_csv("datasets.csv")
ds
```

```python
X = ds[['GRE Score','TOEFL Score','University Rating','SOP','LOR','CGPA','Research']].values
Y = ds['Chance of Admit '].values
Y
```

```python
reg = LinearRegression()
reg.fit(X,Y)
```

```python
pred = reg.predict(X)
pred
```

```python
ls = pd.DataFrame(pred.reshape(len(pred),-1),columns= ['predicted'],index = None)
ls['Actual'] = ds['Chance of Admit ']
ls1 = ls.head(25)
```

```python
ls1.plot(kind='bar',figsize=(10,8))
mb.grid(which='major', linestyle='-', linewidth='0.5', color='green')
mb.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
mb.show()
```
